# 1. memory related terms
#### 1.1 memory:
- alu (arithmetic & logical unit): responsible for operations (e.g. addition/subtraction)
- control unit: decode program instructions (opcodes) read from memory
    - uses additional register called instruction pointer (IP)
    - or program counter (PC) to point currentyl executing instruction
- register: memory location quickly accessible from alu &/or control unit (execution units), extreme fast in terms of access time (no place for data closer to exercution units)
    - accumulators special simplified kind of registers
- word: fixed-base unit of data, e.g. size of most registers, max adress, largest block fo data trasferred in single operation
    - commonly number of bits, wrod size or word length

#### 1.3 compilation
all these languages still must be:
- transformed from text to binary form

- by tool called **compiler** (or assembling tool for assembly tool)
- end result is **binary code format** that may be executed later
    - a sequence of opcodes and their arguments (operands).

#### 1.8 automatic memory management
Mechanism that 
- removes from the programmer the responsibility of manual memory management so that once created, 
- objects are automatically destroyed 
- and the memory after them recovered 
- when no longer needed.

##### 1.8.1 the mutator
the mutator is an entity responsible for executing application code:
- driving machine of all the changes in the application with respect to the memory.
- the state of the memory
- objects are being allocated or modified, and
- references between them are being changed.
- anything that can modify memory, either by updating existing objects or by creating new ones.

##### 1.8.2 mutator: three operations 
- `new(amount)`: allocates amount of memory, to be used by newly created object.
- `write(address, value)`: 
    - writes a specified value 
    - at a given address. 
    - object field (in object-oriented programming), 
    - global variable, 
    - or any other kind of data organization.
- `read(address)`: read value from specified address.

##### 1.8.3 the allocator
provide two main operations:
- `allocate(amount)`: 
    - allocates a specified amount of memory, 
    - internally used by **mutator.new** operation
- `deallocate(address)`: 
    - Frees memory at given address to make it available for future allocations, 
    - internal and not expored to mutator
    - no user code can call it explicitly
    - two common: `sequential` & `free-list`
